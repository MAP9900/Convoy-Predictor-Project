```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler as SS
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
```

```python
file_path = 'heart_disease.csv'
heart_disease_data = pd.read_csv(file_path, delimiter=';')
X = np.array(heart_disease_data[['Age', 'Resting_Blood_Pressure', 'Cholesterol', 'Max Heart Rate', 'ST_depression']])
y = heart_disease_data['site'].values
```

```python
SS = SS()
```

```python
y
```

```python
def K_Fold(model, X, y, K, scaler=None, random_state=146):
    kf = KFold(n_splits=K, random_state=random_state, shuffle=True)
    train_scores = []
    test_scores = []
    for idxTrain, idxTest in kf.split(X):
        Xtrain = X[idxTrain]
        Xtest = X[idxTest]
        ytrain = y[idxTrain]
        ytest = y[idxTest]
        if scaler is not None:
            Xtrain = scaler.fit_transform(Xtrain)
            Xtest = scaler.transform(Xtest)
        model.fit(Xtrain, ytrain)
        train_scores.append(model.score(Xtrain, ytrain))
        test_scores.append(model.score(Xtest, ytest))
    return train_scores, test_scores
```

```python
k_values = np.array(range(10,41))
avg_tr_score = []
avg_te_score = []
for a in k_values:
    knn = KNeighborsClassifier(n_neighbors=a, weights='uniform')
    train_scores, test_scores = K_Fold(knn, X, y, 5, SS)
    avg_tr_score.append(np.mean(train_scores))
    avg_te_score.append(np.mean(test_scores))
n_neighbors = k_values[np.argmax(avg_te_score)]
n_neighbors_2 = k_values[np.argmax(avg_tr_score)]
print('Optimal Number of Neighbors (Test Score):', n_neighbors, np.mean(avg_te_score))
```

```python

file_path = 'heart_disease.csv'
heart_data = pd.read_csv(file_path, delimiter=';')
heart_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

cleveland_data = heart_data[heart_data['site'] == 'Cleveland']

# Selecting the specified features and target
features = cleveland_data[['Age', 'Resting_Blood_Pressure', 'Cholesterol', 'Max Heart Rate', 'ST_depression']]
target = cleveland_data['num']

# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.40, random_state=125, shuffle=True)

# Scaling the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Creating and fitting KNN classifier
knn = KNeighborsClassifier(n_neighbors=19, weights='uniform')
knn.fit(X_train_scaled, y_train)

# Scoring KNN classifier on training and test data
knn_train_score = knn.score(X_train_scaled, y_train)
knn_test_score = knn.score(X_test_scaled, y_test)

# Generating confusion matrices for KNN
knn_confusion_train = confusion_matrix(y_train, knn.predict(X_train_scaled))
knn_confusion_test = confusion_matrix(y_test, knn.predict(X_test_scaled))

# Creating and fitting Logistic Regression model
log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)

# Scoring Logistic Regression on training and test data
log_reg_train_score = log_reg.score(X_train_scaled, y_train)
log_reg_test_score = log_reg.score(X_test_scaled, y_test)

# Generating confusion matrices for Logistic Regression
log_reg_confusion_train = confusion_matrix(y_train, log_reg.predict(X_train_scaled))
log_reg_confusion_test = confusion_matrix(y_test, log_reg.predict(X_test_scaled))

{
    "knn_train_score": knn_train_score,
    "knn_test_score": knn_test_score,
    "knn_confusion_train": knn_confusion_train,
    "knn_confusion_test": knn_confusion_test,
    "log_reg_train_score": log_reg_train_score,
    "log_reg_test_score": log_reg_test_score,
    "log_reg_confusion_train": log_reg_confusion_train,
    "log_reg_confusion_test": log_reg_confusion_test
}
```
